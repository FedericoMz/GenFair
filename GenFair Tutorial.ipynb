```python
import sys, pandas as pd
sys.path.insert(0, "...") #directory of "genfair" folder
from genfair import GenFair
```

```python
german = pd.read_json("/genfair_german_train.json")

german
```

```python
sensitive_attributes = ['Age', 'Sex'] #['sex', 'race']
class_name = 'Risk'


#Experimental, for causal relations. Unused in the paper
#multiple independent features are supported, e.g. [[['Job', 'Credit amount'], 'Housing']]
causal_reg = [] #regressor
causal_class = [] #classifier


integer = [] #Integer  Strategy
weighted = ['Job','Housing', 'Saving accounts', 'Checking account',
       'Credit amount', 'Duration', 'Purpose'] #Weighted Probability  strategy
random_l = [] #Random Probability  strategy

#if a feature is in none of the above: Float Strategy 
```

```python
gf = GenFair(german, sensitive_attributes, class_name, #<- must be included
                               causal_reg, causal_class, #<- causal relations
                               integer, weighted, random_l) #<- how to handle the various features
```

```python
gf.get_results()
```

```python
gf.fit()
```

```python
gf.balance()
```
